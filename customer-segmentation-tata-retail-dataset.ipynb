```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
from sklearn.preprocessing import StandardScaler
```

```python
retail = retail_data
```

```python

# Identify missing values
print("Missing Values:\n", retail.isnull())

# Drop rows with any missing values
retail_dropna = retail.dropna()
print("\nDataFrame after dropping rows with missing values:\n", retail_dropna)
```

```python
retail = retail_dropna
```

```python
retail['CustomerID'] = retail['CustomerID'].astype(str)
```

```python
# Calculate 'Amount' for each transaction
retail['Amount'] = retail['Quantity'] * retail['UnitPrice']

# Calculate Monetary (total spending per customer)
monetary = retail.groupby('CustomerID')['Amount'].sum().reset_index()
monetary.columns = ['CustomerID', 'Monetary']

# Display the Monetary DataFrame
print(monetary)
```

```python
monetary.head(20)
```

```python
monetary.shape
```

```python
monetary.info()
```

```python
retail.head(20)
```

```python
# Calculate Frequency (number of transactions per customer)
frequency = retail.groupby('CustomerID')['InvoiceNo'].count().reset_index()
frequency.columns = ['CustomerID', 'Frequency']

# Display the Frequency DataFrame
print(frequency)
```

```python
frequency.head(20)
```

```python
# Merge the two DataFrames on 'CustomerID'
rfm = pd.merge(monetary, frequency, on='CustomerID',how='inner')

rfm.head(20)
```

```python
# Convert 'InvoiceDate' to datetime format with specific format
retail['InvoiceDate'] = pd.to_datetime(retail['InvoiceDate'], format='%d-%m-%Y %H:%M')
```

```python
retail
```

```python
recent_date = max(retail['InvoiceDate'])
recent_date
```

```python
retail['Difference'] = (recent_date - retail['InvoiceDate']).dt.days
```

```python
retail
```

```python
recency = retail.groupby('CustomerID')['Difference'].min()
recency = recency.reset_index()
recency.head(20)
```

```python
retail.columns
```

```python

final_rfm = pd.merge(rfm,recency,on='CustomerID',how='inner')
final_rfm.columns = ['CustomerID','Amount','Frequency','Recency']
final_rfm.head(20)
```

```python
attributes = ['Amount','Frequency','Recency']
# Set the figure size
plt.rcParams['figure.figsize'] = [10, 8]

# Create the box plot
sns.boxplot(data=final_rfm[attributes], orient="v", palette="Set2", saturation=1, width=0.7)

# Set the title and labels with styling
plt.title("Outliers Variable Distribution", fontsize=14, fontweight='bold')
plt.ylabel("Range", fontweight="bold")
plt.xlabel("Attributes", fontweight="bold")

# Display the plot
plt.show()
```

```python
# Function to remove statistical outliers based on IQR
def remove_outliers(final_rfm, column):
    Q1 = final_rfm[column].quantile(0.05)
    Q3 = final_rfm[column].quantile(0.95)
    IQR = Q3 - Q1
    return final_rfm[(final_rfm[column] >= Q1 - 1.5 * IQR) & (final_rfm[column] <= Q3 + 1.5 * IQR)]

# Remove outliers for Amount, Recency, and Frequency
rfm = remove_outliers(final_rfm, 'Amount')
rfm = remove_outliers(final_rfm, 'Recency')
rfm = remove_outliers(final_rfm, 'Frequency')

# Display the updated DataFrame
rfm
```

```python
rfm_df = rfm[['Amount', 'Frequency', 'Recency']]

# Instantiate
scaler = StandardScaler()

# fit_transform
rfm_df_scaled = scaler.fit_transform(rfm_df)
rfm_df_scaled.shape
```

```python
rfm_df_scaled
```

```python
rfm_df_scaled = pd.DataFrame(rfm_df_scaled)
rfm_df_scaled.columns = ['Amount', 'Frequency', 'Recency']
rfm_df_scaled
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


# Apply K-means clustering
kmeans = KMeans(n_clusters=4, max_iter=50)
kmeans.fit(rfm_df_scaled)
```

```python
kmeans.labels_
```

```python
set(kmeans.labels_)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Assuming rfm_df_scaled is your scaled dataset
# Calculate WCSS for different values of K
wcss = []
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]

for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50, n_init=10, random_state=0)
    kmeans.fit(rfm_df_scaled)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.plot(range_n_clusters, wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Apply K-means clustering
kmeans = KMeans(n_clusters=3, max_iter=300)
kmeans.fit(rfm_df_scaled)
```

```python
import pickle
# Serialize the model using pickle

filename = 'kmeans_model.pkl'

with open('kmeans_model_saved.pkl', 'wb') as file:
    pickle.dump(kmeans, file)
    
file.close()
pickle.dump(kmeans, open('kmeans_model.pkl','wb'))
print("K-means model saved successfully.")
```
